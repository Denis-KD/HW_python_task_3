<h1>Задание 2.1</h1>

На основании данных из задания 1 необходимо составить максимальное(миним 5) кол-во гипотез о корреляциях между отдельными столбцами и проверить их (аналогично воршопу). При этом при составлении гипотез можно рассматривать не полный df, а именно выборки (например: "женщины старше 40 лет покапают в основном осенью")

```python
import pandas as pd

data = pd.read_csv('shopping_trends.csv')

print('1)Гипотеза: Мужчины тратят больше женщин')
print('///////////////////////////////////////////////////////////')
average_purchase_amount_by_gender = data.groupby('Gender')['Purchase Amount (USD)'].mean()

print("Сравнение средних сумм покупок для мужчин и женщин:")
print(average_purchase_amount_by_gender)


print('///////////////////////////////////////////////////////////')
print('2)Гипотеза :  люди от 20 до 30 делают покупки чаще других')
print('///////////////////////////////////////////////////////////')

age_group_below_20 = data[data['Age'] < 20]
age_group_20_to_30 = data[(data['Age'] >= 20) & (data['Age'] <= 30)]
age_group_above_30 = data[data['Age'] > 30]


total_purchases = len(data)

purchases_in_below_20 = len(age_group_below_20)
purchases_in_20_to_30 = len(age_group_20_to_30)
purchases_in_above_30 = len(age_group_above_30)

percentage_purchases_in_below_20 = (purchases_in_below_20 / total_purchases) * 100
percentage_purchases_in_20_to_30 = (purchases_in_20_to_30 / total_purchases) * 100
percentage_purchases_in_above_30 = (purchases_in_above_30 / total_purchases) * 100

print(f"Процент от общего числа покупок для группы до 20 лет: {percentage_purchases_in_below_20:.2f}%")
print(f"Процент от общего числа покупок для группы от 20 до 30 лет: {percentage_purchases_in_20_to_30:.2f}%")
print(f"Процент от общего числа покупок для группы от 30 лет: {percentage_purchases_in_above_30:.2f}%")

print('///////////////////////////////////////////////////////////')
print('3)Гипотеза: те кто покупают ежедневно используют бесплатную доставку ')
print('///////////////////////////////////////////////////////////')

weekly_shoppers = data[(data['Frequency of Purchases'] == 'Weekly') & (data['Shipping Type'] == 'Free Shipping')]

percentage_free_shipping_weekly = (len(weekly_shoppers) / len(data[data['Frequency of Purchases'] == 'Weekly'])) * 100

print(f"Процент заказов с бесплатной доставкой среди покупателей, которые покупают ежедневно: {percentage_free_shipping_weekly:.2f}%")


print('///////////////////////////////////////////////////////////')
print('4) Гипотеза: покупатели, выбирающие "Express" доставки, совершают более дорогие покупки по сравнению с другими методами доставки')
print('///////////////////////////////////////////////////////////')

average_purchase_by_shipping = data.groupby('Shipping Type')['Purchase Amount (USD)'].mean()

print(average_purchase_by_shipping)

print('///////////////////////////////////////////////////////////')
print('5) Гипотеза Продажи летом преобладают товары белого цвета')
print('///////////////////////////////////////////////////////////')

summer_sales = data[data['Season'] == 'Summer']

color_sales_counts = summer_sales['Color'].value_counts()

top_colors_summer = color_sales_counts.head(5)
print(top_colors_summer)
```

<h1>Задание 2.2</h1>

Помимо гипотез, давайте найдем самый популярный товар, построим распределение покупателей по полу, определим какой пол(и отдельно возраст) покупает больше всего, чаще всего, самые дорогие товары. Определим есть ли зависимость между цветом одежды и сезоном. Так же посчитаем наш сезонный mau (уникальных пользователей за сезон) и его динамику. Так же давайте найдем самую популярную букву в названии одежды. 

```python
# самый популярный товар
popular_item = data['Item Purchased'].value_counts().idxmax()
print(f"Самый популярный товар: {popular_item}")
print('///////////////////////////////////////////////////////////')

# распределение покупателей по полу

gender_distribution = data['Gender'].value_counts()
print("Распределение покупателей по полу:")
print(gender_distribution)
print('///////////////////////////////////////////////////////////')

# определим какой пол(и отдельно возраст) покупает больше всего, чаще всего, самые дорогие товары
# Оставим самые дорогие товары
top_expensive_items = data.nlargest(3, 'Purchase Amount (USD)')

# Находим пол и возраст для этой группы товаров

gender_age_data = top_expensive_items[['Gender', 'Age']]

# Определяем пол и возраст, которые чаще всего покупают самые дорогие товары

most_frequent_gender_expensive = gender_age_data['Gender'].value_counts().idxmax()
most_frequent_age_group_expensive = gender_age_data['Age'].apply(
    lambda age: 'До 20 лет' if age < 20 else ('От 20 до 30 лет' if age < 30 else 'От 30 лет')
).value_counts().idxmax()

print("Пол, который чаще всего покупает самые дорогие товары:", most_frequent_gender_expensive)
print("Возрастная группа, которая чаще всего покупает самые дорогие товары:", most_frequent_age_group_expensive)
print('///////////////////////////////////////////////////////////')

#есть ли зависимость между цветом одежды и сезоном

color_season_dependency = data.groupby(['Color', 'Season']).size().unstack(fill_value=0)
print("Зависимость между цветом одежды и сезоном:")
print(color_season_dependency)
print('///////////////////////////////////////////////////////////')

#сезонный mau (уникальных пользователей за сезон) и его динамику
season_mau = data['Season'].unique()

mau_data = []

for i, season in enumerate(season_mau):
    season_data = data[data['Season'] == season]
    
    unique_users = season_data['Customer ID'].nunique()
    
    if i > 0:
        prev_mau = mau_data[i - 1]['MAU']
        mau_change = unique_users - prev_mau
    else:
        mau_change = 0
        
    mau_data.append({'Season': season, 'MAU': unique_users, 'MAU Change': mau_change})

mau_df = pd.DataFrame(mau_data)
print(mau_df)
print('///////////////////////////////////////////////////////////')

# найдем самую популярную букву в названии одежды.
item_names = data['Item Purchased']
letters = [letter for item_name in item_names for letter in item_name if letter.isalpha()]
popular_letter = pd.Series(letters).value_counts().idxmax()
print(f"Самая популярная буква в названии одежды: {popular_letter}")
```

<h1>Задание 4</h1> 

У нас есть Полный каталог всех случаев, когда кто-то проклинал или истекал кровью в фильме Квентина Тарантино. https://github.com/fivethirtyeight/data/tree/master/tarantino

Давайте представим эти данные в виде df и определим:

- число проклятий, летальных исходов и их отношение по фильмам 

- частота употребления конкретных прокрятий 

- распределение времени между проклятиями 

```python
import pandas as pd

df = pd.read_csv('tarantino.csv')


print('1) число проклятий, летальных исходов и их отношение по фильмам')
print('///////////////////////////////////////////////////////////')
films = df['movie'].unique()

curses_count = {}
deaths_count = {}

for film in films:
    curses_count[film] = len(df[(df['movie'] == film) & (df['type'] == 'word')])
    deaths_count[film] = len(df[(df['movie'] == film) & (df['type'] == 'death')])

for film in films:
    curse_ratio = curses_count[film] / deaths_count[film]
    print(f"Фильм: {film}, Число проклятий: {curses_count[film]}, Число летальных исходов: {deaths_count[film]}, Отношение: {curse_ratio}")
print('///////////////////////////////////////////////////////////')

print('2) частота употребления конкретных прокрятий')
print('///////////////////////////////////////////////////////////')
curses = df['word'][df['type'] == 'word']
unique_curses, curse_counts = curses.value_counts().index, curses.value_counts().values

for curse, count in zip(unique_curses, curse_counts):
    print(f"Проклятие: {curse}  {count}")
print('///////////////////////////////////////////////////////////')

print('3) распределение времени между проклятиями')
print('///////////////////////////////////////////////////////////')
time_between_courses_on_films = []

for film in films:
    timestamps = df['minutes_in'][(df['type'] == 'word') & (df['movie'] == film)]
    time_diff = timestamps.diff().dropna()
    
    time_between_courses_on_film = time_diff.mean()
    time_between_courses_on_films.append(time_between_courses_on_film)

time_between_curses = pd.Series(time_between_courses_on_films).mean()
print("Среднее время между проклятиями в каждом фильме:", time_between_courses_on_films)
print("Среднее время между проклятиями по всем фильмам:", time_between_curses)
```
